```
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sn
from mpl_toolkits import mplot3d
import numpy as np
```

```
from google.colab import drive
drive.mount('/content/drive')

save_dir = 'drive/My Drive/adv_img/'
```

```
# mnist 5 epochs
res = np.load(save_dir + 'mnist_result.npz', allow_pickle=True)
res = res['arr_0'].item()


# before
fig = plt.figure()
ax = plt.axes(projection='3d')
n_classes = list(res['before'].keys())
pixels = list(res['before'][n_classes[0]].keys())

for nb_classes in n_classes:
  acc = []
  for ps in pixels:
    acc.append(res['before'][nb_classes][ps])
  ax.plot3D([nb_classes]*len(pixels), pixels, acc, 'gray')

ax.set_xlabel('# classes')
ax.set_ylabel('pixel')
ax.set_zlabel('accuracy')
plt.title('MNIST Acc Before Attack')
plt.show()

# after 
fig = plt.figure()
ax = plt.axes(projection='3d')
n_classes = list(res['after'].keys())
pixels = list(res['after'][n_classes[0]].keys())

for nb_classes in n_classes:
  acc = []
  for ps in pixels:
    acc.append(res['after'][nb_classes][ps])
  ax.plot3D([nb_classes]*len(pixels), pixels, acc, 'gray')


ax.set_xlabel('# classes')
ax.set_ylabel('pixel')
ax.set_zlabel('accuracy')
plt.title('MNIST Acc After Attack')
plt.show()
```

```
#MNIST
x = [14, 23,24,25,26,27,28]
binary = [60,80,62.13,60.37, 78.45, 66.2,69.39]
trinary = [48.2, 46.48, 62.19, 55.42, 46.26, 60.06, 78.88]

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_ylim(40,100)
ax.set_xlim(14,28)
plt.plot(x, binary, 'go-', label='2 class')
for i,j in zip(x,binary):
    ax.annotate(str(j),xy=(i,j))

plt.plot(x, trinary, 'rs-', label='3 class')
for i,j in zip(x,trinary):
    ax.annotate(str(j),xy=(i,j))

plt.legend()
plt.title('MNIST')
plt.xlabel('Pixel')
plt.ylabel('Acc After Attack (%)')
plt.show()

```
